```python
#!/usr/bin/env python3
"""Simple audio test script"""

import requests
# Send audio file to API
chat_response = requests.post(
    f"http://127.0.0.1:8080/chat/",
    json={
        "sessionId": "12r31351341234",
        "inputMessage": "what is the population of riyadh?"
    }
)

# Sample 1: CHAT mode
print("CHAT Response:")
print(json.dumps(chat_response.json(), indent=2))
print("\n" + "="*50 + "\n")
```

```python
import requests
import json

# Base URL for your API
base_url = "http://localhost:8080"

# Sample 1: CHAT mode
chat_response = requests.post(
    f"{base_url}/chat",
    json={
        "sessionId": "CHAT",
        "inputMessage": "Hello there!"
    }
)
print("CHAT Response:")
print(json.dumps(chat_response.json(), indent=2))
print("\n" + "="*50 + "\n")

# Sample 2: CHART mode
chart_response = requests.post(
    f"{base_url}/chat",
    json={
        "sessionId": "CHART",
        "inputMessage": "Show me the population chart"
    }
)
print("CHART Response:")
print(json.dumps(chart_response.json(), indent=2))
print("\n" + "="*50 + "\n")

# Sample 3: MAP mode
map_response = requests.post(
    f"{base_url}/chat",
    json={
        "sessionId": "MAP",
        "inputMessage": "Display regional population data"
    }
)
print("MAP Response:")
print(json.dumps(map_response.json(), indent=2))
print("\n" + "="*50 + "\n")

# Sample 4: Custom session ID (will get default response)
custom_response = requests.post(
    f"{base_url}/chat",
    json={
        "sessionId": "CUSTOM123",
        "inputMessage": "This is a custom session"
    }
)
print("Custom Session Response:")
print(json.dumps(custom_response.json(), indent=2))
```
